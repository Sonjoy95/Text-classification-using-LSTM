```python
import tensorflow.compat.v2
```

```python
from keras.preprocessing.text import Tokenizer
```

```python
from keras.preprocessing import sequence
```

**Tokenization:** Texts are splits into words and words are known as tokens, the process of spliting the text into words are known as Tokenization.

In Keras we have **texts_to_sequences** function which is used to split the texts into a lists of words.

```python
max_words=1000
max_length=150
tok=Tokenizer(num_words=max_words)
tok.fit_on_texts(x_train)
sequences=tok.texts_to_sequences(x_train)
sequences_matrix=sequence.pad_sequences(sequences,maxlen=max_length)
```

```python
import numpy as np
from keras.models import Model
from keras.layers import LSTM,Activation,Dense,Dropout,Input,Embedding
from tensorflow.keras.utils import to_categorical
from keras.callbacks import EarlyStopping
```

**RNN model**

**Embedding layer:** In Keras we have Embedding layer that can be used for neural networks on text data. It requires that the input data be integer encoded, so that each word is represented by a unique integer.

```python
def RNN():
    inputs=Input(name="inputs",shape=[max_length])
    layer1=Embedding(max_words,50,input_length=max_length)(inputs)
    layer2=LSTM(64)(layer1)
    layer3=Dense(256,name="FC1")(layer2)
    layer4=Activation("relu")(layer3)
    layer5=Dropout(0.5)(layer4)
    layer6=Dense(1,name="out_layer")(layer5)
    layer7=Activation("sigmoid")(layer6)
    model=Model(inputs=inputs,outputs=layer7)
    return model
```

```python
model=RNN()
model.summary()
```

**Compile and fit**

```python
model.compile(loss="binary_crossentropy",optimizer="RMSprop",metrics=["accuracy"])
```

```python
rnn_model=model.fit(sequences_matrix,y_train,batch_size=128,epochs=10,validation_split=0.2
        )
```

**Accuracy**

```python
accuracy_df=pd.DataFrame({"Train_accuracy":rnn_model.history["accuracy"],
                          "val_accuracy":rnn_model.history["val_accuracy"]})
accuracy_df
```

**Function to predict ham or spam**

```python
def predict(context):
    max_len=150
    txt=tok.texts_to_sequences(context)
    txt=sequence.pad_sequences(txt,maxlen=max_len)
    pred=model.predict(txt)
    if np.round(pred)==0:
        print("it is a Ham")
    else:
        print("It is spam")
```
